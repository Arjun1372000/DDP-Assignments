## Q. Explain the concept of distributed data processing. How does Apache Spark improve performance compared to traditional data processing systems? Include the role of in-memory computation.

Distributed data processing involves splitting large datasets into smaller parts and processing them across multiple machines in parallel. This improves scalability, fault tolerance, and processing speed by utilizing cluster resources efficiently.

### How Apache Spark Improves Performance

Apache Spark enhances performance compared to traditional systems like Hadoop MapReduce through:

- **In-memory computation**: Stores intermediate data in RAM, reducing disk I/O  
- **Lazy evaluation**: Builds an execution plan (DAG) before running tasks, enabling optimization  
- **Reduced disk usage**: Avoids writing intermediate results to disk after every step  
- **Efficient scheduling**: Uses a DAG scheduler to optimize task execution  

### Role of In-Memory Computation

In-memory computation allows Spark to cache data in RAM, making repeated and iterative operations much faster. This is especially useful for machine learning and data analytics tasks where the same data is processed multiple times.

### Summary

Distributed processing enables parallel computation on large datasets. Spark improves performance mainly through in-memory processing and optimized execution, making it significantly faster than traditional systems.

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("BasicDF").getOrCreate()
```

```python
data = [
    (1, "Arjun", 22, 25000),
    (2, "Rahul", 25, 35000),
    (3, "Sneha", 28, 40000),
    (4, "Priya", 24, 28000),
    (5, "Kiran", 30, 50000)
]
```

```python
columns = ["ID", "Name", "Age", "Salary"]
```

```python
df = spark.createDataFrame(data, columns)
```

```python
df.printSchema()
```

```python
df.show()
```
