### **Q. What is Structured Streaming in Spark? Explain how it differs from batch processing. Discuss its key features and real-time use cases.**

Structured Streaming is a scalable stream processing engine built on Spark SQL.  
It allows processing of real-time data using the same APIs as batch processing by treating streams as **unbounded tables**.

---

### Difference: Streaming vs Batch Processing

| Aspect            | Batch Processing                     | Structured Streaming                  |
|------------------|-------------------------------------|--------------------------------------|
| Data             | Finite (stored data)                | Infinite (continuous data stream)    |
| Execution        | Runs once on complete dataset       | Runs continuously                    |
| Latency          | High (delayed results)              | Low (near real-time results)         |
| Use case         | Reports, analytics                  | Live dashboards, monitoring          |

---

### Key Features
- Unified API for batch and streaming  
- Fault tolerance using checkpointing  
- Exactly-once processing guarantees  
- Automatic optimization using Spark engine  

---

### Real-time Use Cases
- Log and event monitoring  
- Fraud detection in transactions  
- Live analytics dashboards  
- IoT sensor data processing

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, split
import os
import time
```

```python
# Start Spark session
spark = SparkSession.builder.appName("FileStream").getOrCreate()
```

```python
# Create a folder for streaming data
input_path = "stream_data"
os.makedirs(input_path, exist_ok=True)
```

```python
# Read stream from the folder
# Spark will process new files added to this folder as a stream
lines = spark.readStream.format("text").load(input_path)
```

```python
# Split lines into words
words = lines.select(explode(split(lines.value, " ")).alias("word"))
```

```python
# Word count
word_counts = words.groupBy("word").count()
```

```python
# Write output to console
query = word_counts.writeStream \
    .outputMode("complete") \
    .format("console") \
    .start()
```
